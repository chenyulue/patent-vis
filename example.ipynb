```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

fig, ax = plt.subplots()
ax.set_xlim([0, 100])
ax.set_ylim([0, 100])

total1 = sq['一级'].loc[:, '_area_'].sum()
total2 = sq['二级'].loc[:, '_area_'].sum()

for rect, area in zip(sq['一级'].loc[:, '_rect_'], sq['一级'].loc[:, '_area_']):
    #print('一级: ', rect['dx']*rect['dy']/10000, area/total1)
    #if abs(rect['dx']*rect['dy']/10000 - area/total1) < 1e-4:
        rect1 = mpatches.Rectangle((rect['x'], rect['y']), rect['dx'], rect['dy'], ec='w', fc='yellow')
        ax.add_patch(rect1)
    
for rect, area in zip(sq['二级'].loc[:, '_rect_'], sq['二级'].loc[:, '_area_']):
    #print('二级: ', rect['dx']*rect['dy']/10000, area/total1)
    #if abs(rect['dx']*rect['dy']/10000 - area/total1) < 1e-4:
        rect1 = mpatches.Rectangle((rect['x'], rect['y']), rect['dx'], rect['dy'], ec='w', fc='orange')
        ax.add_patch(rect1)
    
for rect, area in zip(sq['三级'].loc[:, '_rect_'], sq['三级'].loc[:, '_area_']):
    #print('三级: ', rect['dx']*rect['dy']/10000, area/total1)
    #if abs(rect['dx']*rect['dy']/10000 - area/total1) < 1e-4:
        rect1 = mpatches.Rectangle((rect['x'], rect['y']), rect['dx'], rect['dy'], ec='w', fc='green')
        ax.add_patch(rect1)


for rect, area in zip(sq['四级'].loc[:, '_rect_'], sq['四级'].loc[:, '_area_']):
    #print('四级: ', rect['dx']*rect['dy']/10000, area/total1)
    #if abs(rect['dx']*rect['dy']/10000 - area/total1) < 1e-4:
        rect1 = mpatches.Rectangle((rect['x'], rect['y']), rect['dx'], rect['dy'], ec='w')
        ax.add_patch(rect1)
    
plt.show()
```
